MELT()

```python
import pandas as pd
df=pd.DataFrame({
    'days':[1,2,3,4,5,6,7],
    'visitors':[103,99,100,81,77,86,97],
    'bounce_rate':[65,67,78,65,66,85,100],
    'week_s_day':['Sun','Sun','Mon','Mon','Tue','Tue','Wed']
})
print(df)
```

```python
pd.melt(df)
```

```python
pd.melt(df,id_vars=['bounce_rate']) #bounce_rate work as id
```

```python
pd.melt(df,id_vars=['bounce_rate'],var_name='visitors') #visitors work as value
```

#pivot Function

```python
import pandas as pd
df=pd.DataFrame({
    'days':[1,2,3,4,5,6,7],
    'visitors':[103,99,100,81,77,86,97],
    'bounce_rate':[65,67,78,65,66,85,100],
    'week_s_day':['Sun','Sun','Mon','Mon','Tue','Tue','Wed']
})
print(df)
```

```python
df.pivot(index='days',columns='week_s_day')
```

```python
df.pivot(index='days',columns='week_s_day',values='visitors') #it provide value of visitors per day
```

```python
df.pivot_table(index='days',columns='week_s_day',aggfunc="mean") #it provide value of bounce_rate per day
```
