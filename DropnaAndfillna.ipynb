#Handling Missing Data by Dropena and fillna

```python
import pandas as pd
```

```python
data=pd.read_excel("/content/DATA.xlsx")
data
```

```python
data.dropna() #it has removed all the row with nan data
```

```python
data.dropna(axis=1) #it has removed all the column with nan data
```

```python
data.dropna(how="all") #it has removed all the row with all nan data
```

```python
data.dropna(subset=["Hindi"]) #it has removed all the nan value of hindi column
```

```python
data.dropna(inplace=True) #update dataframe with removing nan value provide a new data
data
```

```python
data.dropna(thresh=1) #drop the row which has only one/2 null value if we put 2 inplace of 1
data
```

#Fillna

```python
data=pd.read_excel("/content/DATA.xlsx")
data
```

```python
data.fillna("Python")#it fill all the nan value with python
```

```python
data.fillna({'Name':'Rohit','Roll no':9})#it filled nan data from that coulun with particular value like here in name column filled nan value with rohit
```

```python
data.fillna(method="ffill") #it filled next nan data with previous one
```

```python
data.fillna(method="bfill") #it filled previous nan data with next one
```

```python
data.fillna(method="bfill",axis=1)
```

```python
data.fillna(12,inplace=True) #it filled null value with 12
data
```

```python
data.fillna("Python",limit=3) #it filled 3 nan value with python
data
```
