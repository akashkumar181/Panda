```python
import pandas as pd
```

```python
var=pd.DataFrame({
    'Name':["Rahul","Mohit","Gita","Sita","Rahul","Mohit","Gita","Sita"],
    'Subject1':[34,89,56,89,34,89,56,89],
    'Subject2':[34,89,59,89,34,89,96,89]
})
var

```

```python
varNew=var.groupby('Name')
varNew
```

```python
for x, y in varNew:
  print(x)
  print(y)
```

```python
varNew.get_group('Rahul')
```

```python
varNew.min()
```

```python
varNew.mean()
```
