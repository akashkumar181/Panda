```python
import pandas as pd
```

```python
var=pd.DataFrame({"A":[1,2,3],"B":[4,5,6],"C":[7,8,9]})
var
```

Addition

```python
var["total of A+B+C"]=var["A"]+var["B"]+var["C"]
var
```

Subtract

```python
var["total of A+B+C"]=var["A"]+var["B"]-var["C"]
var
```

Multiplication

```python
var["total of A+B+C"]=var["A"]*var["B"]*var["C"]
var
```
