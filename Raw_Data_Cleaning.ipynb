```python
import pandas as pd
```

```python
data=pd.read_csv("/content/sales.csv")
data
```

Handle Missing Values:

Identify any missing or NaN values in the dataset.

Fill missing values in PaymentMethod with 'Unknown'.

For ShippingStatus, correct any typos (like "Pendin") to "Pending".

For DiscountApplied, replace any missing values with "No".


```python
data.isna().sum()
```

```python
data["ShippingStatus"]=data["ShippingStatus"].replace("Pendin","Pending")
data
```

```python
data.head(100)
```

```python
data["DiscountApplied"]=data["DiscountApplied"].fillna("No") #filled missing value with nan value
data
```

```python
data["PaymentMethod"]=data["PaymentMethod"].fillna("Unknown")
data
```

```python
data.head(100)
```

Remove Duplicates:

Check for and remove any duplicate rows from the dataset.

Correct Data Types:

Ensure the Date column is in a datetime format.

Verify that numerical columns like Price and Quantity are of the correct data type.

Filter Invalid Entries:

Remove any rows where Quantity or Price are negative

```python
data.drop_duplicates() #use to remove duplicates
```

```python
missing_dates=data[data["Date"].isna()]
print(missing_dates)
```

```python
data["Date"]=data["Date"].fillna(method="ffill")


data
```

```python
delivered_count=(data['ShippingStatus']=='Delivered').sum()
pending_count=(data['ShippingStatus']=='Pending').sum()
shipped_count=(data['ShippingStatus']=='Shipped').sum()
print("Delivered_count",delivered_count)
print("Pending_count",pending_count)
print("Shipped_count",shipped_count)
```

```python
Status_count=pd.DataFrame({
    'Shipping Status':['Delivered','Pending','Shipped'],
    'Count':[delivered_count,pending_count,shipped_count],


})
Status_count
```
