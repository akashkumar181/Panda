Write csv file

```python
import pandas as pd
dis={
    'A':[1,2,3,4],
    'B':[1,2,3,4],
    'C':[1,2,3,4],
    'D':[1,2,3,4]
}
data=pd.DataFrame(dis)
x=data.to_csv('test1.csv',index=False)#index =false to remove index
x                         #creating a csv file of data
```

```python
dis={
    'A':[1,2,3,4],
    'B':[1,2,3,4],
    'C':[1,2,3,4],
    'D':[1,2,3,4]
}
data=pd.DataFrame(dis)
x=data.to_csv('test2.csv',index=False,header=["AA","BB","CC","DD"])#CHANGING THE HEADER NAME IT MEANS CHANGING THE COLUMN NAME OF EXCEL SHEET
x
```
